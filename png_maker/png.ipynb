```python
from PIL import Image
import numpy as np

# Load image
img = Image.open("initial_image.png").convert("RGBA")

# Convert to numpy array
data = np.array(img)

# Define threshold for "white/grey"
threshold = 200  # increase if some grey remains

# Separate channels
r, g, b, a = data.T

# Condition for white & grey pixels
white_or_grey = (r > threshold) & (g > threshold) & (b > threshold)

# Set alpha to 0 (fully transparent) for those pixels
data[..., 3][white_or_grey.T] = 0

# Convert back to image
transparent_img = Image.fromarray(data)

# Save result
transparent_img.save("logo_transparent.png")

print("✅ Transparent PNG saved as logo_transparent.png")
```
